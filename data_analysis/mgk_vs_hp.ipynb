```python
from analysis_utils import *
import os
```

```python
directory_paths = {
    '22_HP': "../22_HP/outputs",
    '22_MGK': "../22_MGK/outputs",
    '32_HP': "../32_HP/outputs",
    '32_MGK': "../32_MGK/outputs",
    '45_HP': "../45_HP/outputs",
    '45_MGK': "../45_MGK/outputs"
}
```

```python
dfs = {}
```

```python
for node, directory_path in directory_paths.items():
    # List all files in the directory
    file_names = os.listdir(directory_path)
    
    # Filter CSV files
    csv_files = [file for file in file_names if file.endswith('nand.csv')]
    
    # Initialize DataFrame
    df = pd.DataFrame()
    
    # Concatenate data from all CSV files
    for file_name in csv_files:
        df = pd.concat([df, pd.read_csv(f"{directory_path}/{file_name}")])
    
    dfs[node] = df
    
```

```python
def leak_vs_temp(nodes, colors):
    # Create the plot with subplots
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    for i, node in enumerate(nodes):
        df_node = dfs[node]
        axs[i].scatter(df_node['temp'], df_node['leakage'], color=colors[i], label=node)
        axs[i].set_xlabel('Temperature')
        axs[i].set_ylabel('Leakage')
        axs[i].set_title(f'{node} - Leakage vs Temperature')
        axs[i].legend()
        axs[i].grid(True)

    # Set overall title
    fig.suptitle('Leakage vs Temperature Comparison')

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
```

```python
nodes = [('22_HP', '22_MGK'), ('32_HP', '32_MGK'), ('45_HP', '45_MGK')]
colors = ['red', 'blue']
for node in nodes:
    leak_vs_temp(node, colors)
```

```python
nodes = [('22_HP', '22_MGK'), ('32_HP', '32_MGK'), ('45_HP', '45_MGK')]
```

```python
colors = ['red', 'blue']
for node, color in zip(nodes, colors):
    leak_vs_temp(node, color)
```
